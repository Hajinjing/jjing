#apply()

```python
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','fare']] #열이름이 age와 fare인 
df['ten'] = 10
print(df.head())
```

```python
def add_5(n):
  return n+5
```

```python
for i in range(len(df)):
  df.iloc[i, 0] = add_5(df.iloc[i, 0])
print(df.age)
```

```python
sr1 = df['age'].apply(lambda x: x*3-4)
print(sr1)
```

```python
df1 = pd.DataFrame({'a':['a0','a1','a2','a3'],
                    'b':['b0','b1','b2','b3'],
                    'c':['c0','c1','c2','c3']},
                    index=[0,1,2,3])
print(df1)
```

```python
df2 = pd.DataFrame({'a':['a2','a3','a4','a5'],
                    'b':['b2','b3','b4','b5'],
                    'c':['c2','c3','c4','c5'],
                    'd':['d2','d3','d4','d5']},
                    index=[2,3,4,5])
print(df2)
```

```python
result1 = pd.concat([df1,df2])
print(result1)
```

```python
result1 = pd.concat([df1,df2], ignore_index=True)
print(result1)
```

```python
result1 = pd.concat([df1,df2])
result1.reset_index(drop=True,inplace=True)
print(result1)
```

```python
result2 = pd.concat([df1,df2], axis='columns')
print(result2)
```

```python
result3 = pd.concat([df1,df2], axis='columns', join='inner')  #교집합 인덱스가 겹치는
print(result3)
```

```python
df = titanic.loc[:,['age','sex','class','fare','survived']]
print(df.head())
```

```python
df['class'].unique()
df['class'].value_counts()
```

```python
grouped = df.groupby(['class'])
print(grouped)
```

```python
grouped_df={}
for key, group in grouped:
  print('key:', key)
  print('length:', len(group))
  grouped_df[key]=group
  print(group.head())
```

```python
print(grouped_df['First'])
```

```python
average = grouped.mean()
print(average) #그룹바이 한 경우에는 그룹별로 평균,최소,최대 등을 구할 수 있음
```

```python
grouped_two = df.groupby(['class','sex'])

for key, group in grouped_two:
  print('key:', key)
  print('length:', len(group))
  print(group.head())
```

```python
print(grouped_two.mean())
```

```python
group3f = grouped_two.get_group(('Third', 'female'))
print(group3f.head())
```

```python
dfg = grouped_two.mean()
print(dfg)
```

```python
print(dfg.loc[('First','male')])
```

```python
print(dfg.loc['First'])
```

```python
print(dfg.xs('female', level='sex'))
```

```python
pdf1 = pd.pivot_table(df,
                      index='class',
                      columns='sex',
                      values='age',
                      aggfunc='mean')
print(pdf1)
print(type(pdf1))
```

```python
pdf2 = pd.pivot_table(df,
                      index='class',
                      columns='sex',
                      values=['age','fare'],
                      aggfunc='mean')
print(pdf2)
print(type(pdf2))
```

```python
pdf3 = pd.pivot_table(df,
                      index='class',
                      columns='sex',
                      values='age',
                      aggfunc=['mean', 'sum','std'])
print(pdf3)
print(type(pdf3))
```

```python
pdf4 = pd.pivot_table(df,
                      index=['class', 'sex'],
                      columns='sex',
                      values='age',
                      aggfunc='mean')
print(pdf4)
print(type(pdf4))
```
