```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi']=70
```

```python
fig = plt.figure(figsize=(8,5))
ax=fig.add_axes([0,0,1,1])   #비율임
ax.set_title('figure')
plt.show()
```

```python
X = [0,1]
Y = [0,1]
fig = plt.figure(figsize=(8,5))
ax = [None, None, None]
ax[0] = fig.add_axes([0.1,0.1,0.4,0.8])
ax[1] = fig.add_axes([0.55,0.15,0.35,0.4])
ax[2] = fig.add_axes([0.65,0.6,0.2,0.3])
ax[0].plot(X,Y)
for i in range(3):
  ax[i].set_title('ax[{}]'.format(i))
  ax[i].set_xticks([])  #빈 리스트를 주면 좌표값이 안그려짐
  ax[i].set_yticks([])
plt.show()
```

```python
fig, axes = plt.subplots(2,3, figsize=(8,5))
axes[1,2].plot(X,Y)
plt.tight_layout()  #ax끼리 간격 조절
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))  #bpi 기준으로   560픽셀
axes=[]
axes.append(plt.subplot2grid((3,4), (0,0), colspan=2, rowspan=2))
axes.append(plt.subplot2grid((3,4), (0,2)))
axes.append(plt.subplot2grid((3,4), (0,3), rowspan=3))
axes.append(plt.subplot2grid((3,4), (1,2)))
axes.append(plt.subplot2grid((3,4), (2,2)))
axes.append(plt.subplot2grid((3,4), (2,0), colspan=3))
for i in range(6):
  axes[i].set_title('axes[{}]'.format(i))
  axes[i].set_xticks([])
  axes[i].set_yticks([])
fig.tight_layout(pad=2)
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))  #bpi 기준으로   560픽셀
gs = fig.add_gridspec(3,4)
axes = [None]*5
axes[0] = fig.add_subplot(gs[0:2, 0:2])
axes[1] = fig.add_subplot(gs[0,2])
axes[2] = fig.add_subplot(gs[:,-1])
axes[3] = fig.add_subplot(gs[1,2])
axes[4] = fig.add_subplot(gs[2,:-1])
for i in range(5):
  axes[i].set_title('axes[{}]'.format(i))
  axes[i].set_xticks([])
  axes[i].set_yticks([])
fig.tight_layout(pad=2)
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
gs = fig.add_gridspec(3,4)  #그리드는 3행4열

axes = [None]*5
axes[0] = fig.add_subplot(gs[0:2, 0:2])  #0~1행 0~1열
axes[4] = fig.add_subplot(gs[2,3]) #2행의 3열 칸에 만들어지고
```

```python
fig, axes = plt.subplots()
axin1 = axes.inset_axes([0.8, 0.1, 0.15, 0.15], xticks=[], yticks=[])
plt.show
```

```python
from mpl_toolkits.axes_grid1 import axes_divider
from mpl_toolkits.axes_grid1.axes_divider import make_axes_locatable

fig, axes = plt.subplots()    #하나의 fig안에 여러 plot들을 표현
ax_divider = make_axes_locatable(axes)
axes = ax_divider.append_axes('right', size='10%', pad='5%', xticks=[], yticks=[])
plt.show()
```
