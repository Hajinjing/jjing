```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
training_data = np.array(
    [[0,0],[0,1],[1,0],[1,1]], 'float32')

target_data = np.array([[0],[1],[1],[0]], 'float32')
```

```python
model = Sequential()
model.add(Dense(32, input_dim=2, activation='relu'))  #두개의 값을 입력 받아서 32개의 dense로 연결  히든레이어
model.add(Dense(1, activation='sigmoid'))  #출력레이어
model.compile(loss='mse', optimizer='adam',
              metrics=['binary_accuracy'])

print(model.summary())
```

```python
fit_hist = model.fit(training_data, target_data, epochs=500, verbose=1)
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```

```python
plt.plot(fit_hist.history['binary_accuracy'])
plt.show()
```

```python
inp = list(map(int, input('논리값을 입력하세요:').split()))
qwe = np.array(inp) #input 값을 array타입으로 바꿔서
print('입력 값 : ', qwe)
qwe = qwe.reshape(-1,2)  #행개수는 아무렇게나 만들어라 열은 2개
print('reshape:', qwe)
print('결과 값:', model.predict(qwe)[0][0].round())
```
