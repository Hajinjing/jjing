```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('/content/drive/MyDrive/datasets/wholesale.csv')
print(data.head())
```

```python
scaled_data = normalize(data)
scaled_data = pd.DataFrame(scaled_data, columns=data.columns)
print(scaled_data.head())
```

```python
plt.figure(figsize=(5,4))
plt.title('Dendrograms')
shc.dendrogram(shc.linkage(scaled_data, method='ward'))
plt.axhline(y=4, color='r', linestyle='--')
plt.show()
```

```python
cluster = KMeans(n_clusters = 3)
print(cluster.fit_predict(scaled_data))
```

```python
plt.figure(figsize=(5, 4))
plt.scatter(scaled_data['Milk'],
            scaled_data['Grocery'],
            c=cluster.labels_)
plt.show()
```

```python
pd.plotting.scatter_matrix(scaled_data, figsize=(15, 15),
                           c=cluster.labels_)
plt.show()
```
